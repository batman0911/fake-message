```python
import random
import pandas as pd
```

```python
NUM_CUSTOMER = 1000000
START_YEAR = 1960
END_YEAR = 2000


user_data = []

random.seed(94)

for i in range(NUM_CUSTOMER):
  gender = 1 if random.random() < 0.5 else 0
  user_data.append([i, gender, random.randint(START_YEAR, END_YEAR)])
```

```python
user_data
```

```python
df = pd.DataFrame(user_data, columns=['id', 'gender', 'birth_year'])
```

```python
df
```

```python
df.to_csv('../data/user-data.csv', index=False)
```
